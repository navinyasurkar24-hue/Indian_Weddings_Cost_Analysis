## Average wedding cost across different Wedding Type categories.

```python
wedding_cost=df.groupby('Wedding_Type')['Cost_of_Type'].mean()

plt.figure(figsize=(18,8))
sb.barplot(x=wedding_cost.index,y=wedding_cost.values)
plt.xlabel('Wedding Type')
plt.ylabel('Average Cost of Wedding')
plt.title('Average Wedding Cost by Wedding Type')
plt.show()
```

### Heritage Weddings, Palace Weddings, and Cruise Weddings have the highest average costs, indicating they are premium or luxury weddings.
### Destination Weddings, Beach Weddings, Farmhouse Weddings, and Resort Weddings have comparatively lower average costs, making them more cost-efficient or budget-friendly options.

```python
plt.figure(figsize=(20,8))
sb.lineplot(x=wedding_cost.index,y=wedding_cost.values)
plt.xlabel('Wedding Type')
plt.ylabel('Average Cost of Wedding')
plt.title('Trend of Average Wedding Cost Across Wedding Types')
plt.show()
```

### The line chart highlights a clear cost trend where luxury wedding types drive higher average expenses, while other wedding types remain comparatively budget-friendly.

```python
plt.figure(figsize=(8,8))
plt.pie(wedding_cost.values,labels=wedding_cost.index,autopct='%d%%')
plt.title('Distribution of Average Wedding Cost by Wedding Type')
plt.show()
```

### Heritage, Palace, and Cruise Weddings occupy larger portions of the pie, indicating they contribute a higher share to overall wedding spending and represent the luxury segment.

### Destination, Beach, Farmhouse, and Resort Weddings have smaller slices, reflecting their lower contribution to total average costs and positioning them as more budget-friendly options.

```python
decor_cost=df.groupby('Decor_Category')['Decor'].mean()
entertainment_cost=df.groupby('Entertainment_Category')['Entertainment'].mean()
gifts_cost=df.groupby('Gifts_Category')['Gifts(per_piece)'].mean()

plt.figure(figsize=(20,5))

plt.subplot(1,3,1)
sb.barplot(x=decor_cost.index,y=decor_cost.values)
plt.xlabel('Decor Category')
plt.ylabel('Average Decor Cost')
plt.title('Average Decor Cost by Category')
plt.xticks(rotation=45)

plt.subplot(1,3,2)
sb.barplot(x=entertainment_cost.index,y=entertainment_cost.values)
plt.xlabel('Entertainment Category')
plt.ylabel('Average Entertainment Cost')
plt.title('Average Entertainment Cost by Category')
plt.xticks(rotation=45)

plt.subplot(1,3,3)
sb.barplot(x=gifts_cost.index,y=gifts_cost.values)
plt.xlabel('Gifts Category')
plt.ylabel('Average Gifts Cost (Per Piece)')
plt.title('Average Gifts Cost by Category')
plt.xticks(rotation=45)

plt.show()
```
